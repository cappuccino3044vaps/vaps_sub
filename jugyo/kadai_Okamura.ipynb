$s_a(t) = A \cos (2 \pi f t + \phi)$


```python
import matplotlib.pyplot as plt
import numpy as np
#from google.colab import files

#####################################
## Time Domain Representation
#####################################
# Time sequences
t = np.linspace(0,2.55,256)
# Amplitude
A = 1
# Constant Frequency
fc = 7
# Initial Phase
phi = 0

fig_a = plt.figure(figsize=(15,5))

ax_a=plt.subplot(121)
sa = A*np.cos(2*np.pi*fc*t + phi)
ax_a.plot(t, sa)
ax_a.set_xlabel('Time (s)')
ax_a.set_ylabel('Amplitude')
ax_a.set_title('Time domain representation')
####################################
## Frequency Domain Representation
####################################
ax_a2=plt.subplot(122)
f = np.linspace(0,100,256)

Sa = np.fft.fft(sa)
Saa=np.abs(Sa)
ax_a2.plot(f,Saa)
ax_a2.set_xlim(0,50)
ax_a2.set_xlabel('Frequency (Hz)')
ax_a2.set_ylabel('Amplitude Spectrum')
ax_a2.set_title('Frequency domain representation')

```
