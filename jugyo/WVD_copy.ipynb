```python
#######
import numpy as np
import matplotlib.pyplot as plt
#from google.colab import files
from scipy import signal

#####################################
##  Signal Parameters
#####################################
T=255     # 
N=256     # 
dt=1       # 

# 
f0 = 0.1  # 
fe = 0.2  # 

a = (fe - f0)/T  # Gradient parameter

# 
t = np.arange(0,N,dt)
A=1

## 
s =A*np.cos(2*np.pi*(f0*t+(a/2)*t**2))


fig_a = plt.figure(figsize=(10,2))
ax_a = fig_a.add_subplot()
ax_a.plot(t, s)
ax_a.set_xlabel('Time (s)')
ax_a.set_ylabel('Amplitude')
ax_a.set_title('Signal')


s0=np.zeros(int(N/2))
tmp=np.concatenate([s0, s],0)
es=np.concatenate([tmp,s0],0)

###########################
# Make the signal kernel
##########################

Kz = np.empty((N,0),complex)
for tt in range(N):
  #real signal
  s1 = es[tt:tt+N]
  s2 = np.flip(np.conjugate(s1))
  S = s1*s2
  Kz = np.append(Kz,S.reshape(N,1),axis=1)

tt = 200
s1 = es[tt:tt+N]
s2 = np.flip(np.conjugate(s1))
S = s1*s2

#
fig_b = plt.figure(figsize=(10,6))
ax_b1 = plt.subplot(311)
ax_b1.plot(s1)
ax_b1.set_ylabel('s1')
ax_b1.set_xlabel(' ')
ax_b2 = plt.subplot(312)
ax_b2.plot(s2)
ax_b2.set_ylabel('s2')
ax_b2.set_xlabel(' ')
ax_b3 = plt.subplot(313)
ax_b3.plot(S)
ax_b3.set_ylabel('s1*s2')
ax_b3.set_xlabel('tau')

###############
#WVD
###############

Ws = np.fft.fft(Kz,axis=0)
WD = abs(Ws.T)
WD_roll=np.roll(WD,int(N/2),1)

#
fig=plt.figure(figsize=(10,10))
ax_c=fig.add_subplot()
plt.imshow(WD_roll,cmap='jet',origin='lower'); plt.colorbar()
ticks = ax_c.set_xticks([25, 76, 128, 178, 229])
labels = ax_c.set_xticklabels(['-0.2', '-0.1', '0', '0.1', '0.2'])
ax_c.set_xlabel('Frequency (Hz)')
ax_c.set_ylabel('Time (s)')
ax_c.set_title('WVD of real chirp signal')

```
