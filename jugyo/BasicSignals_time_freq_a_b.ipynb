1. $s_a(t) = A \cos (2 \pi f t + \phi)$


```
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

#####################################
## Time Domain Representation
#####################################
# Time sequences
t = np.linspace(0,2.55,256)
# Amplitude
A = 1
# Constant Frequency
fc = 7
# Initial Phase
phi = 0

fig_a = plt.figure(figsize=(15,5))

ax_a=plt.subplot(121)
sa = A*np.cos(2*np.pi*fc*t + phi)
ax_a.plot(t, sa)
ax_a.set_xlabel('Time (s)')
ax_a.set_ylabel('Amplitude')
ax_a.set_title('Time domain representation')

####################################
## Frequency Domain Representation
####################################
ax_a2=plt.subplot(122)
f = np.linspace(0,100,256)

Sa = np.fft.fft(sa)
Saa=np.abs(Sa)
ax_a2.plot(f,Saa)
ax_a2.set_xlim(0,50)
ax_a2.set_xlabel('Frequency (Hz)')
ax_a2.set_ylabel('Amplitude Spectrum')
ax_a2.set_title('Frequency domain representation')
```

2. ${\displaystyle s_b(t)= \sum_{k=1}^{M} \cos (2 \pi f_{k} t + \phi_{k})}$

```
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

#####################################
## Time Domain Representation
#####################################
# Time sequences
t = np.linspace(0,2.55,256)
# Multiple amplitudes
A = np.array([1.0, 0.9, 0.7])
# Multiple Frequencies
f = np.array([5.0, 7.0, 9.0])
# Initial Phases
phi = np.array([2, 0.0, 3.0])

fig_b = plt.figure(figsize=(15,5))

ax_b=plt.subplot(121)
sb = np.zeros(256)
for i in (0, 1, 2):
  sb += A[i]*np.cos(2*np.pi*f[i]*t + phi[i])
ax_b.plot(t, sb)
ax_b.set_xlabel('Time (s)')
ax_b.set_ylabel('Amplitude')
ax_b.set_title('Time domain representation')

####################################
## Frequency Domain Representation
####################################
ax_b2=plt.subplot(122)
f = np.linspace(0,100,256)

Sb = np.fft.fft(sb)
Sba=np.abs(Sb)
ax_b2.plot(f,Sba)
ax_b2.set_xlim(0,50)
ax_b2.set_xlabel('Frequency (Hz)')
ax_b2.set_ylabel('Amplitude Spectrum')
ax_b2.set_title('Frequency domain representation')
```
